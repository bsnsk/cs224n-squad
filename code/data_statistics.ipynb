```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def plot_histogram(filename):
    with open("../data/train.%s" % filename) as f:
        lens = [len(line.split(" ")) for line in f]
    #print Counter(lens)
    plt.figure()
    plt.hist(lens, bins=50)
    plt.xlabel('%s length' % filename)
    plt.ylabel('Occurrence')
    plt.title('Histogram of length of %s' % filename)
    plt.grid(True)
    plt.show()
    plt.savefig("../statistics/" + filename + "_hist.eps")
    
filenames = ["question", "context", "answer"]
for filename in filenames:
    plot_histogram(filename)

        
```

```python
with open('../data/train.context') as f1, open('../data/train.answer') as f2:
    end_of_ans = []
    for line1, line2 in zip(f1, f2):
        index = line1.find(line2)
        if index is not -1:
            cnts = len(line1[:index-1].split(" "))
            cnts += len(line2.split(" "))
            end_of_ans.append(cnts)
    plt.hist(end_of_ans, bins=50)
    plt.xlabel('length')
    plt.ylabel('Occurrence')
    plt.title('Histogram of end of ans in context')
    plt.grid(True)
    plt.show()
    plt.savefig("../statistics/end_of_ans_hist.eps")
            
        
```
